```python
# number of points
n_points = [1000, 5000, 10000, 15000, 30000, 50000]

# create line plot
mpi_line, = plt.plot(n_points, mpi_times, color='blue', label='MPI')
openmp_line, = plt.plot(n_points, openmp_times, color='red', label='OpenMP')
hybrid_line, = plt.plot(n_points, hybrid_times, color='green', label='Hybrid')

# add highlights for given number of points
highlight_points = [1000, 5000, 10000, 15000, 30000, 50000]
for point in highlight_points:
    index = n_points.index(point)
    time_mpi = mpi_times[index]
    time_openmp = openmp_times[index]
    time_hybrid = hybrid_times[index]
    plt.plot(point, time_mpi, 'o', color=mpi_line.get_color())
    plt.plot(point, time_openmp, 'o', color=openmp_line.get_color())
    plt.plot(point, time_hybrid, 'o', color=hybrid_line.get_color())

# set plot title and axis labels
plt.title('Algorithm time vs. Number of Points')
plt.xlabel('Number of Points')
plt.ylabel('Time (s)')

# add legend
plt.legend()

# display plot
plt.show()
```
